# Gr√°fico de barras: Poblaci√≥n por edad

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../csv/titanicnuevo.csv")
df = df.loc[0:890]

poblacionPorEdad = {
    "0-9": 0,
    "10-19": 0,
    "20-29": 0,
    "30-39": 0,
    "40-49": 0,
    "50-59": 0,
    "60-69": 0,
    "70-79": 0,
    "80-89": 0,
    "90-99": 0,
    "100+": 0
}

def obtenerPoblacionPorEdad(valor):
  global i
  global poblacionPorEdad
  if i != 100 and valor >= i-10 and valor < i:
    poblacionPorEdad[f"{i-10}-{i-1}"] += 1
  elif i == 100 and valor >= i:
    poblacionPorEdad[f"{i}+"] += 1

for i in range(10, 110, 10):
  df["Age"].apply(obtenerPoblacionPorEdad)

plt.bar(poblacionPorEdad.keys(), poblacionPorEdad.values())
plt.show()
```

Este gr√°fico de barras muestra una cantidad aproximada de pasajeros pertenecientes a distintos rangos de edad (expresados en a√±os). Se utiliza como eje X distintos rangos de edad y como eje Y la cantidad de pasajeros que encajan en dicho rango. Se puede ver claramente que la mayor√≠a de los pasajeros eran relativamente j√≥venes.

# Gr√°fico de torta: V√≠ctimas y supervivientes

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../csv/titanicnuevo.csv")
df = df.loc[0:890]

def obtenerSupervivientes(valor):
  global victimas
  global supervivientes
  if valor == 0:
    victimas += 1
  else:
    supervivientes += 1

victimas = 0
supervivientes = 0
df["Survived"].apply(obtenerSupervivientes)

labels = ["Sobrevivi√≥", "No sobrevivi√≥"]
sizes = [supervivientes, victimas]
colors = ["green", "red"]
fig, ax = plt.subplots(figsize=(5, 5))

ax.pie(sizes, labels=labels, autopct='%.2f%%',startangle=90, colors=colors, pctdistance=0.5)
plt.title("¬øCu√°l era la probabilidad de sobrevivir al hundimiento del Titanic?")
plt.show()
```

Este gr√°fico de torta muestra la proporci√≥n de v√≠ctimas y supervivientes del hundimiento del barco. Se concluye que la posibilidad era relativamente baja, menos de la mitad de los pasajeros sobrevivieron al accidente.
