# Gráfico de puntos: Precio de las tarifas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../csv/titanicnuevo.csv")
df = df.loc[0:890]

valoresX = df["PassengerId"]
valoresY = df["Fare"]
plt.scatter(valoresX, valoresY)
plt.show()
```

Este gráfico de puntos utiliza como eje X el ID de cada pasajero y como eje Y el precio de la tarifa que pagó. Se puede concluir que la inmensa mayoría de los pasajeros pagaron tarifas por debajo de 100 libras.

# Gráfico de barras: Población por edad

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../csv/titanicnuevo.csv")
df = df.loc[0:890]

poblacionPorEdad = {
    "0-9": 0,
    "10-19": 0,
    "20-29": 0,
    "30-39": 0,
    "40-49": 0,
    "50-59": 0,
    "60-69": 0,
    "70-79": 0,
    "80-89": 0,
    "90-99": 0,
    "100+": 0
}

def obtenerPoblacionPorEdad(valor):
  global i
  global poblacionPorEdad
  if i != 100 and valor >= i-10 and valor < i:
    poblacionPorEdad[f"{i-10}-{i-1}"] += 1
  elif i == 100 and valor >= i:
    poblacionPorEdad[f"{i}+"] += 1

for i in range(10, 110, 10):
  df["Age"].apply(obtenerPoblacionPorEdad)

plt.bar(poblacionPorEdad.keys(), poblacionPorEdad.values())
plt.show()
```

Este gráfico de barras muestra una cantidad aproximada de pasajeros pertenecientes a distintos rangos de edad (expresados en años). Se utiliza como eje X distintos rangos de edad y como eje Y la cantidad de pasajeros que encajan en dicho rango. Se puede ver claramente que la mayoría de los pasajeros eran relativamente jóvenes.

# Gráfico de torta: Víctimas y supervivientes

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../csv/titanicnuevo.csv")
df = df.loc[0:890]

def obtenerSupervivientes(valor):
  global victimas
  global supervivientes
  if valor == 0:
    victimas += 1
  else:
    supervivientes += 1

victimas = 0
supervivientes = 0
df["Survived"].apply(obtenerSupervivientes)

labels = ["Sobrevivió", "No sobrevivió"]
sizes = [supervivientes, victimas]
colors = ["green", "red"]
fig, ax = plt.subplots(figsize=(5, 5))

ax.pie(sizes, labels=labels, autopct='%.2f%%',startangle=90, colors=colors, pctdistance=0.5)
plt.title("¿Cuál era la probabilidad de sobrevivir al hundimiento del Titanic?")
plt.show()
```

Este gráfico de torta muestra la proporción de víctimas y supervivientes del hundimiento del barco. Se concluye que la posibilidad era relativamente baja, menos de la mitad de los pasajeros sobrevivieron al accidente.
